3d

```python
from matplotlib import cm

fig = plt.figure()
ax = fig.gca(projection='3d')
X = np.arange(-5, 5, 0.25)
Y = np.arange(-5, 5, 0.25)
X, Y = np.meshgrid(X, Y)
R = np.sqrt(X**2 + Y**2)
Z = np.sin(R)
surf = ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap=cm.coolwarm)
plt.show()
```

## subplot
여러개의 그래프를 그릴 수 있다.

subplot(nrows, ncols, index, **kwargs)

* nrows: 가로줄이 몇개
* ncols: 세로줄이 몇개

index는 1부터 시작

$ y_1 = \cos(x_1) \cdot e^{-x_1} $

$ y_2 = \frac {1} {1 + e^{-x_2}} $

* tight_layout: Adjust the padding between and around subplots.
  그래프 사이 간격을 조절한다.

```python
x1 = np.linspace(0, 5)
x2 = np.linspace(-8, 8)
y1 = np.cos(2 * np.pi * x1) * np.exp(-x1)
y2 = 1 / (1 + np.exp(-x2))

plt.subplot(2, 2, 1)
plt.title("1st")
plt.plot(x1, y1, 'y')

plt.subplot(2, 2, 2)
plt.hist(np.random.randn(1000), bins=30)

plt.subplot(2, 2, 3)
plt.title("2nd")
plt.plot(x2, y2, 'r')

plt.tight_layout()
plt.show()
```

## subplots
그래프가 들어갈 자리들을 동시에 생성

```python
_, axes = plt.subplots(4, 4)
for ax in axes.flat:
    ax.plot(np.random.rand(10))
plt.tight_layout()
plt.show()
```
